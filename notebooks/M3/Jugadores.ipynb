```python
import pandas as pd
import numpy as np
```

```python
#1 Crea un DataFrame con Datos de Jugadores
datos = {
    "Jugador": ["Lionel Messi", "Cristiano Ronaldo", "Kevin De Bruyne", "Kylian Mbappe", "Luka Modric"],
    "Posicion": ["Delantero", "Delantero", "Mediocampista", "Delantero", "Mediocampista"],
    "Edad":[35,38,31,24,37],
    "Goles":[20,18,8,25,3],
    "Asistencias":[10,5,15,12,8]
}

df = pd.DataFrame(datos)

print("DataFrame:")
print(df.head())
```

```python
#2 Muestra los diferentes Jugadores
print("\n Jugadores:")
print(df["Jugador"].unique())
```

```python
#3 Filtra por nombre y cantidad de goles (>10)
suma_goles = df.groupby("Jugador")["Goles"].sum()
goleadores = suma_goles[suma_goles > 10]
print(goleadores)
```

```python
#4 Agrega columna 'Puntos' con Goles*4 - 'Asistencias'*2
df.insert(5, "Puntos", (df["Goles"] * 4) + (df["Asistencias"] * 2))
print(df)
```

```python
#5 Promedio de goles de todos los jugadores
promedio_goles = df["Goles"].mean()
print(f"El promedio de goles de todos los jugadores es: {promedio_goles}")
```

```python
#6 Maxima y minima asistencia
max_asistencias = df["Asistencias"].max()
min_asistencias = df["Asistencias"].min()
print(f"La maxima asistencia es: {max_asistencias}")
print(f"La minima asistencia es: {min_asistencias}")
#
```

```python
#7 Cuenta cuántos jugadores hay por posición (Delantero, Mediocampista)
posiciones = df.groupby("Posicion")["Jugador"].count()
print(posiciones)
```

```python
#8 Ordena el DataFrame en función de los goles en orden descendente
df.sort_values(by="Goles", ascending=False)
print(df)
```

```python
#9 Aplica describe() para obtener estadísticas generales del DataFrame
df.describe()
```

```python
#10 Usa value_counts() para contar cuántos jugadores hay en cada posición
df["Posicion"].value_counts()
```
