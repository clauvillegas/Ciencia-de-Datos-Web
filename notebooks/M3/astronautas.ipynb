```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)  # Para reproducibilidad

# Generar datos aleatorios simulados
n = 20  # número de astronautas

horas_ejercicio = np.round(np.random.uniform(1, 3, n), 2)  # entre 1 y 3 horas diarias
consumo_calorias = np.round(np.random.normal(2900, 150, n), 0)  # media 2900, desviación 150
niveles_estres = np.random.choice(["Bajo", "Medio", "Alto"], size=n, p=[0.4, 0.4, 0.2])
horas_recreativas = np.random.randint(3, 12, n)  # entre 3 y 11 horas semanales

# Crear DataFrame
df = pd.DataFrame({
    "Horas de ejercicio diario (gravedad cero)": horas_ejercicio,
    "Consumo diario de calorías": consumo_calorias,
    "Nivel de estrés": niveles_estres,
    "Horas recreativas semanales": horas_recreativas
})

# 1. Tabla de distribución: horas recreativas semanales vs nivel de estrés
tabla_distribucion = pd.crosstab(df["Horas recreativas semanales"], df["Nivel de estrés"])
print("Tabla de distribución: Horas recreativas semanales vs Nivel de estrés")
print(tabla_distribucion)

# 2. Gráfico de dispersión: horas de ejercicio diario vs consumo diario de calorías
plt.scatter(df["Horas de ejercicio diario (gravedad cero)"], df["Consumo diario de calorías"], color='purple')
plt.xlabel("Horas de ejercicio diario (gravedad cero)")
plt.ylabel("Consumo diario de calorías")
plt.title("Relación entre ejercicio diario y consumo de calorías")
plt.grid(True)
plt.show()

# 3. Correlación entre horas de ejercicio diario y consumo de calorías
correlacion = df["Horas de ejercicio diario (gravedad cero)"].corr(df["Consumo diario de calorías"])
print(f"Coeficiente de correlación de Pearson: {correlacion:.2f}")
```
