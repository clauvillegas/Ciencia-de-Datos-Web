# **Importar Librerias**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
from google.colab import files
```

```python
files.upload()
archivo_csv = 'datos_trafico.csv'

#leer archivo .csv
with open(archivo_csv, 'r', encoding='utf-8') as f:
    sample = f.read(2048)
    dialect = csv.Sniffer().sniff(sample)

df = pd.read_csv(archivo_csv, sep=dialect.delimiter)
print(df)
```

```python
plt.plot(df['Día'], df['Vehículos'], label='Vehículos', color='purple',  linestyle='--')
plt.xlabel('Día')
plt.ylabel('Cantidad de vehículos')
plt.title('Cantidad de vehículos por día')
plt.legend()
plt.show()
```

```python

plt.hist(df['Velocidad_Promedio'], bins=8, histtype='bar',  rwidth=0.8 , color='purple' )
plt.xlabel('Velocidad (km/h)')
plt.ylabel('Frecuencia')
plt.title('Distribución de velocidades de los vehículos')
plt.show()

```

```python

plt.scatter(df['Vehículos'], df['Accidentes'], label='Accidentes')
plt.xlabel('Vehículos')
plt.ylabel('Accidentes')
plt.title('Relación entre vehículos y accidentes')
plt.legend()
plt.xticks(rotation=45)
plt.savefig('grafico_accidentes.png')
plt.show()
```
