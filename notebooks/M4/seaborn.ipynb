```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creacion Dataset
df = pd.DataFrame({
    'Edad': np.arange(25, 75, 5),
    'Ingresos': np.arange(50000, 150000, 10000) + np.random.normal(0, 5000, 10),
    'Años_Educación': np.arange(12, 32, 2) + np.random.normal(0, 1, 10),
    'Horas_Sueño': [7, 6.5, 7.5, 8, 7, 6, 7, 8, 7.5, 6.5]

})

```

```python

print(df)

#Matriz Correlacion
matriz_correlacion = df.corr()

#Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', linewidths=2)
plt.title('Matriz de Correlación')
plt.tight_layout()
#girar labels
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
#
```

Se infiere que:

1.   Los ingresos estan fuertemente correlacionados con la edad y los años de educación y no tanto con las horas de Sueño.

2.   La correlacion de las Horas de Sueño con el resto de las varibles es prácticamente nula.
